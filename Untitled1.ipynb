```python
correation_matrix = new_df.corr()
fig,ax = plt.subplots(figsize=(10,6))
ax = sns.heatmap(data=correation_matrix,
                annot=True,
                linewidths=0.45,
                fmt="0.2f",
                cmap="YlGnBu")
```

### The above heatmap shows us that the yearly amount spent is most dependent on length of membership

```python
fig, ax = plt.subplots(figsize=(35,15))
ax.scatter(new_df["Length of Membership"],new_df["Yearly Amount Spent"],c="green")
ax.scatter(new_df["Time on App"],new_df["Yearly Amount Spent"],c="red")
ax.scatter(new_df["Avg. Session Length"],new_df["Yearly Amount Spent"],c="black")
ax.scatter(new_df["Time on Website"],new_df["Yearly Amount Spent"],c="blue")
;
```

### A very good relationship can be seen with length of membership and yearly amount spent

# Modelling

```python
# Splitting the data into X and y
X = new_df.drop("Yearly Amount Spent", axis=1)
y = new_df["Yearly Amount Spent"]
```

```python
# Splitting it into train and test
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X, y ,test_size=0.2,random_state=0)
```

```python
y_train
```

```python
from sklearn.linear_model import LinearRegression
```

```python
linearreg = LinearRegression()
```

```python
linearreg.fit(X_train, y_train)
```

```python
y_preds = linearreg.predict(X_test)
y_preds
```

```python
y_test
```

```python
linearreg.score(X_test, y_test)
```

```python
fig, ax= plt.subplots(figsize=(10,6))
ax.scatter(y_test, y_preds);
```
